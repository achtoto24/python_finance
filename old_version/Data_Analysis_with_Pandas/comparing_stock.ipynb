```python
import yfinance as yf

# AAPL 데이터 가져오기
aapl = yf.download('AAPL', start='2024-08-01')
# PLTR 데이터 가져오기
pltr = yf.download('PLTR', start='2024-08-01')

#print(nvda.head(10))
aapl = aapl.drop(columns='Adj Close')
print(aapl.tail(10))
print(aapl.index)
print(aapl.columns)
```

```python
#주가 단순 비교 
import matplotlib.pyplot as plt

plt.plot(aapl.index, aapl.Close, 'g', label = 'AAPL')
plt.plot(pltr.index, pltr.Close, 'black', label = 'PALANTIR')
plt.legend(loc = 'best')
```

```python
#pltr['Close']   #one of the Series
#pltr['Close'].shift(1)
print(pltr['Close']); print(); print(pltr['Close'].shift(1))
```

```python
pltr_rate = (pltr['Close'] / pltr['Close'].shift(1) - 1) * 100
pltr_rate.iloc[0] = 0
pltr_rate
```

```python
#일간 변동률 누적곱 
aapl_dpc = (aapl['Close'] - aapl['Close'].shift(1)) / aapl['Close'].shift(1) * 100
aapl_dpc.iloc[0] = 0
aapl_dpc_cp = ((100 + aapl_dpc) / 100).cumprod() * 100 - 100    #일간 변동률 누적곱 계산

pltr_dpc = (pltr['Close'] / pltr['Close'].shift(1) - 1) * 100
pltr_dpc.iloc[0] = 0
pltr_dpc_cp = ((100 + pltr_dpc) / 100).cumprod() * 100 - 100    #일간 변동률 누적곱 계산

plt.plot(aapl.index, aapl_dpc_cp, 'b', label = "APPLE")
plt.plot(pltr.index, pltr_dpc_cp, 'r--', label = "PALANTIR")
plt.xlabel('Period')
plt.ylabel('Change %')
plt.grid(True)
plt.legend(loc = 'best')
plt.show()
```

```python
#최대 손실 낙폭 
nasdaq = yf.download('^IXIC', start='2000-01-01')

window  = 252
peak = nasdaq['Adj Close'].rolling(window, min_periods = 1).max()
drawdown = nasdaq['Adj Close']/ peak - 1.0
max_dd = drawdown.rolling(window, min_periods = 1).min()

plt.figure(figsize = (15, 15))
plt.subplot(211)
nasdaq['Close'].plot(label = "NASDAQ", title = "NASDAQ MDD", grid = True, legend = True)
plt.subplot(212)
drawdown.plot(c = 'blue', label = 'NASDAQ DD', grid = True, legend = True)
max_dd.plot(c = 'red', label = 'NASDAQ MDD', grid = True, legend = True)
plt.show()
```

```python
#지수화 비교 
import yfinance as yf
import matplotlib.pyplot as plt

nasdaq = yf.download('^IXIC', start='2000-01-04')
kospi = yf.download('^KS11', '2000-01-04')

n = (nasdaq['Close'] / nasdaq['Close'].loc['2000-01-04']) * 100
k = (kospi.Close / kospi.Close.loc['2000-01-04']) * 100

plt.figure(figsize = (9, 9))
plt.plot(n.index, n, 'r--', label = 'Nasdaq Average')
plt.plot(k.index, k, 'b', label = 'Kospi Average')
plt.grid(True)
plt.legend(loc = 'best')
plt.show()
```

```python
#산점도 분석
import pandas as pd
df = pd.DataFrame({'NASDAQ' : nasdaq['Close'], 'KOSPI' : kospi['Close']})
df = df.bfill()
df = df.ffill()

plt.figure(figsize = (10, 10))
plt.scatter(df['NASDAQ'], df['KOSPI'], marker= '.')
plt.xlabel('NASDAQ Average')
plt.ylabel('KOSPI Average')
plt.show()

# 점의 분포가 y = x인 직선 형태에 가까울수록 직접적인 관계에 있다. 
```

```python
import pandas as pd
import yfinance as yf
from scipy import stats
import matplotlib.pyplot as plt

nasdaq = yf.download('^IXIC', start='2000-01-04')
kospi = yf.download('^KS11', start = '2000-01-04')

df = pd.DataFrame({'X' : nasdaq["Close"], 'Y' : kospi["Close"]})
df = df.bfill()
df = df.ffill()

regr = stats.linregress(df.X, df.Y)
regr_line = f"Y = {regr.slope : .2f} * X + {regr.intercept: .2f}"

plt.figure(figsize = (10, 10))
plt.plot(df.X, df.Y, '.')
plt.plot(df.X, regr.slope * df.X + regr.intercept, c =  'r')
plt.legend(['NASDAQ X KOSPI', regr_line])
plt.title(f"NASDAQ X KOSPI (R = {regr.rvalue:.2f})")
plt.xlabel("NASDAQ Average")
plt.ylabel("KOSPI Average")
plt.show()

''' <상관계수에 따른 리스크 완화; 현대 포트폴리오 이론 (해리 마코위츠)>
+1.0 : 리스크 완화 효과가 없음
+0.5 : 중간 정도의 리스크 완화 효과가 있음
0    : 상당한 리스크 완화 효과가 있음
-0.5 : 대부분의 리스크를 제거함
-1.0 : 모든 리스크 제거함
'''
```
