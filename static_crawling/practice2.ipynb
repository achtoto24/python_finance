```python
import requests as rq
from bs4 import BeautifulSoup

url = "https://news.naver.com/breakingnews/section/101/259"
data = rq.get(url)
html = BeautifulSoup(data.content)
print(html)
```

```python
html_select = html.select("div.sa_item_flex > div.sa_text > a")
html_select2 = html.select("div.sa_item_flex > div.sa_text > a > strong.sa_text_strong")
print(html_select[0]["href"])
print(html_select2[0].text)
print()
print([i.text for i in html_select2])
```

```python
# POST fromat
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd

url = "https://kind.krx.co.kr/disclosure/todaydisclosure.do"

payload = {
    "method: searchTodayDisclosureSub",
    "currentPageSize: 15",
    "pageIndex: 1",
    "orderMode: 0",
    "orderStat: D",
    "marketType: 1",
    "forward: todaydisclosure_sub",
    "chose: S",
    "todayFlag: N",
    "selDate: 2024-10-08" 
}

data = rq.post(url, data = payload)
print(data)
print()
html = BeautifulSoup(data.content)
print(html)
print()
html_unicode = html.prettify()  # BeautifulSoup에서 파싱한 파서 트리를 다시 유니코드 형태로 돌려줌
print(html_unicode)
```

```python
tbl = pd.read_html(html_unicode)
print(tbl)
```
