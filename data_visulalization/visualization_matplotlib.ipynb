```python
import seaborn as sns
import matplotlib.pyplot as plt # matplotlib : package / pyplot : module

df = sns.load_dataset("penguins")

# scatter()

plt.scatter(df["flipper_length_mm"], df["body_mass_g"])
plt.show()
```

```python
df_group = df.groupby("species")["body_mass_g"].mean().reset_index()

# bar()
plt.bar(x = df_group["species"], height = df_group["body_mass_g"])
plt.show()
```

```python
plt.rc("font", family = "Malgun Gothic")    # 한글 폰트 입력 
plt.hist(df["body_mass_g"], bins = 30)  # bins 인자에는 히스토그램을 몇 개의 구간으로 나눌지를 입력함
plt.xlabel("Body Mass") 
plt.ylabel("Count")
plt.title("팽귄의 몸무게 분포")
plt.show()
```

```python
import pandas as pd

df_unrate = pd.read_csv(
    "https://research.stlouisfed.org/fred2/series/UNRATE/downloaddata/UNRATE.csv"
)

# plot() : line graph

df_unrate["DATE"] = pd.to_datetime(df_unrate["DATE"])   
plt.plot(df_unrate["DATE"], df_unrate["VALUE"]) 
plt.show()
```

```python
# draw many graphs at once

#1. sateless way

fig, axes = plt.subplots(2, 1, figsize = (10, 6))

axes[0].scatter(df["flipper_length_mm"], df["body_mass_g"])
axes[0].set_xlabel("날개 길이(mm)")
axes[0].set_ylabel("몸무게(g)")
axes[0].set_title("날개와 몸무게 간의 관계")

axes[1].hist(df["body_mass_g"], bins = 30)
axes[1].set_xlabel("Body Mass")
axes[1].set_ylabel("Count")
axes[1].set_title("펭귄의 몸무게 분포")

# 간격 조정
plt.subplots_adjust(left = 0.1,
                    right = 0.95,
                    bottom = 0.1,
                    top = 0.95,
                    wspace = 0.5,
                    hspace = 0.5)
plt.show()

'''stateless 방법은 figure 내에 원하는 만큼 axes 객체를 생성하여 그래프들을 그림'''
```

```python
#2 stateful way

plt.figure(figsize = (10, 6))

plt.subplot(2, 1, 1)
plt.scatter(df["flipper_length_mm"], df["body_mass_g"])
plt.xlabel("날개 길이(mm)")
plt.ylabel("몸무게(g)")
plt.title("날개와 몸무게 간의 관계")

plt.subplot(2, 1, 2)
plt.hist(df["body_mass_g"], bins = 30)
plt.xlabel("Body Mass")
plt.ylabel("Count")
plt.title("펭귄의 몸무게 분포")

plt.subplots_adjust(left = 0.1,
                    right = 0.95,
                    bottom = 0.1,
                    top = 0.95,
                    wspace = 0.5,
                    hspace = 0.5)

plt.show()

'''stateful 방법은 subplot()을 이용함'''
```
