```python
import seaborn as sns
import matplotlib.pyplot as plt # matplotlib : package / pyplot : module

df = sns.load_dataset("penguins")

# scatter()

plt.scatter(df["flipper_length_mm"], df["body_mass_g"])
plt.show()
```

```python
df_group = df.groupby("species")["body_mass_g"].mean().reset_index()

# bar()
plt.bar(x = df_group["species"], height = df_group["body_mass_g"])
plt.show()
```

```python
plt.rc("font", family = "Malgun Gothic")    #    
plt.hist(df["body_mass_g"], bins = 30)  # bins       
plt.xlabel("Body Mass") 
plt.ylabel("Count")
plt.title("  ")
plt.show()
```

```python
import pandas as pd

df_unrate = pd.read_csv(
    "https://research.stlouisfed.org/fred2/series/UNRATE/downloaddata/UNRATE.csv"
)

# plot() : line graph

df_unrate["DATE"] = pd.to_datetime(df_unrate["DATE"])   
plt.plot(df_unrate["DATE"], df_unrate["VALUE"]) 
plt.show()
```

```python
# draw many graphs at once

#1. sateless way

fig, axes = plt.subplots(2, 1, figsize = (10, 6))

axes[0].scatter(df["flipper_length_mm"], df["body_mass_g"])
axes[0].set_xlabel(" (mm)")
axes[0].set_ylabel("(g)")
axes[0].set_title("   ")

axes[1].hist(df["body_mass_g"], bins = 30)
axes[1].set_xlabel("Body Mass")
axes[1].set_ylabel("Count")
axes[1].set_title("  ")

#  
plt.subplots_adjust(left = 0.1,
                    right = 0.95,
                    bottom = 0.1,
                    top = 0.95,
                    wspace = 0.5,
                    hspace = 0.5)
plt.show()

'''stateless  figure    axes    '''
```

```python
#2 stateful way

plt.figure(figsize = (10, 6))

plt.subplot(2, 1, 1)
plt.scatter(df["flipper_length_mm"], df["body_mass_g"])
plt.xlabel(" (mm)")
plt.ylabel("(g)")
plt.title("   ")

plt.subplot(2, 1, 2)
plt.hist(df["body_mass_g"], bins = 30)
plt.xlabel("Body Mass")
plt.ylabel("Count")
plt.title("  ")

plt.subplots_adjust(left = 0.1,
                    right = 0.95,
                    bottom = 0.1,
                    top = 0.95,
                    wspace = 0.5,
                    hspace = 0.5)

plt.show()

'''stateful  subplot() '''
```
