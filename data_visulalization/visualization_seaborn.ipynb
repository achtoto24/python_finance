```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.scatterplot(data = df, x = "age", y = "fare")
plt.show()
```

```python
sns.scatterplot(data = df, x = "age", y = "fare", hue = "class", style = "class")   # hue : color, style : shape
plt.show()
```

```python
df_pivot = df.pivot_table(index = "class",
                          columns = "sex",
                          values = "survived",
                          aggfunc = "mean")
df_pivot
```

```python
sns.heatmap(df_pivot, annot = True, cmap = "coolwarm")  # annot :    , cmap :   
plt.show()
```

```python
# draw many graphs at once

#figure-level
sns.displot(data = df, x = "age", hue = "class", kind = "hist", alpha = 0.3)    #alpha : transparency
plt.show()
```

```python
sns.displot(data = df, x = "age", col = "class", kind = "hist")
plt.show()
```

```python
sns.displot(data = df, x = "age", col = "class", row = "sex", kind = "hist")
plt.show()
```

```python
# axes-level

plt.rc("font", family = "Malgun Gothic")

g, axes = plt.subplots(2, 1, figsize = (8, 6))

sns.histplot(data = df, x = "age", hue = "class", ax = axes[0])
sns.barplot(data = df, x = "class", y = "age", ax = axes[1])

axes[0].set_title("  ")
axes[1].set_title("  ")

g.tight_layout()    #  
plt.show()
```
