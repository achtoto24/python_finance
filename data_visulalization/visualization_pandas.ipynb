```python
import seaborn as sns

df = sns.load_dataset("diamonds")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.rc("font", family = "Malgun Gothic")
df.plot.scatter(x = "carat", y = "price", figsize = (10, 6), title = "   ")
plt.show()
```

```python
df.plot.scatter(x = "carat", y = "price", c = "cut", cmap = "Set2", figsize = (10, 6), title = "   ") # c      , cmap  
plt.show()
```

```python
df["price"].plot.hist(figsize = (10, 6), bins = 20)
plt.show()
```

```python
df.groupby("color")["carat"].mean().plot.bar(figsize = (10, 6))
plt.show()
```
