```python
import seaborn as sns

df = sns.load_dataset("penguins")
df.head()
```

```python
# split the group with single column

df_group = df.groupby(["species"])
df_group.head(3)
```

```python
for key, group in df_group :
    print(key)
    display(group.head(2))
```

```python
df_group[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].mean()
```

```python
# split the group with mutiple columns

df.groupby(["species", "sex"])[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].mean()
```

```python
# agg()
# making user function to apply to the group object

def min_max(x) :
    return x.max() - x.min()

df.groupby(["species"])["bill_length_mm"].agg(min_max)
```

```python
df.groupby(["species"])[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].agg(["min", "max"])
```

```python
df.groupby(["species"]).agg({"bill_length_mm" : ["max", "min"], "island" : "count"})
```

```python
# transfprm()

print(df.groupby(["species"])["bill_length_mm"].transform("mean"))
```

```python
def z_score(x) :
    z = (x  - x.mean()) / x.std()
    return (z)

df.groupby(["species"])[["bill_length_mm", "flipper_length_mm"]].transform(z_score)
```

```python
# apply()

df.groupby(["species"])["bill_length_mm"].apply(min)
```

```python
df.groupby(["species"])[["bill_length_mm", "flipper_length_mm"]].apply(z_score)
```

```python
df.groupby(["species"]).filter(lambda x : x["bill_length_mm"].mean() >= 40)
```
