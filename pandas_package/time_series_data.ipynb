```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("taxis")
df.head(10)
```

```python
df.info()   # pickup과 dropoff 열 타입은 object(?), 즉 문자열이다
```

```python
# object 객체를 datetime 객체로 바꾸기 

df["pickup"] = pd.to_datetime(df["pickup"])
df["dropoff"] = pd.to_datetime(df["dropoff"])

df.info()
```

```python
df["pickup"][0].month   # [0]은 첫 번째 행을 의미
```

```python
df["year"] = df["pickup"].dt.year
df["month"] = df["pickup"].dt.month
df["day"] = df["pickup"].dt.day

df[["pickup", "year", "month", "day"]].head()
```

```python
df.sort_values("pickup", inplace = True)        # sort_values는 "pickup"열을 기준으로 데이터 오름차순 정렬
df.reset_index(drop = True, inplace = True)     # reset_index는 행 인덱스 초기화

df.head()
```

```python
df["dropoff"] - df["pickup"]    # datetime 객체이므로 시간 연산 가능
```

```python
df.set_index("pickup", inplace = True)

print(df.head())
print()
print(df.index)
```

```python
print(df.loc["2019-02"])
print()
print(df.loc["2019-03-01" : "2019-03-02"])
```
